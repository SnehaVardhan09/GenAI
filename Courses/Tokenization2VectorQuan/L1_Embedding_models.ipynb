```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
model
```

* Choosing a proper embedding for your data.
* Selecting a model trained on acedamic data but your data is coming from twitter then it impacts the search quality.
* Word based tokenization - Subword tokenization
* **walker walked a long walk** -->  **[10, 46, 26,36, 98]** --> **[[0.10, 1.46...],[2.42,0.36...],[0.30, 0.14,....],[1.52,0.66....]]**

```python
tokenized_data = model.tokenize(["tokenization and vector quantization"])
tokenized_data
```

```python
model.tokenizer.convert_ids_to_tokens(tokenized_data["input_ids"][0])
```

```python
first_module = model._first_module()
first_module.auto_model
```

```python
embeddings = first_module.auto_model.embeddings
embeddings
```

```python
import torch
import plotly.express as px

first_sentence = "vecotr search optimization"
second_sentence = "tokenization ids and embedding"

with torch.no_grad():
  first_tokens = model.tokenize([first_sentence])
  second_tokens = model.tokenize([second_sentence])

  first_embedding = embeddings.word_embeddings(
      first_tokens['input_ids']
  )
  second_embedding = embeddings.word_embeddings(
      second_tokens['input_ids']
  )
first_embedding.shape, second_embedding.shape
```

```python
from sentence_transformers import util

distances = util.cos_sim(
    first_embedding.squeeze(),
    second_embedding.squeeze()
).cpu().numpy() # Move the tensor to the CPU and convert to a NumPy array

px.imshow(
    distances,
    x=model.tokenizer.convert_ids_to_tokens(
        second_tokens["input_ids"][0]
    ),
    y=model.tokenizer.convert_ids_to_tokens(
        first_tokens["input_ids"][0]
    ),
    text_auto=True,
)
```
